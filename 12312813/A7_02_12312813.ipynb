````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
# we are using 2019 data 
df = pd.read_csv("2019.csv")
# display basic information about the dataset
print(df.info())
````

````python
# whether there is null values 
print(df.isnull().sum())
# there is no need to dropna
````

````python
# Select only numeric columns for correlation matrix
numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns
correlation_matrix = df[numeric_columns].corr()

# Plot the correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
````

* Firstly, point out the analysis method and list the results

  From the correlation analysis of the heatmap above, it can be seen that the three most relevant variables to the happiness score (`Score`) (excluding Score itself) are `GDP per capita` (an indicator of economic development level), `Healthy life expectancy` and `Social support` (the sum of behaviors in which a certain social network uses certain material and spiritual means to provide free assistance to vulnerable groups in society). The correlation coefficients between these three variables and the happiness score are 0.79, 0.78, and 0.78, respectively, indicating a strong positive correlation between them and the happiness score.

* Provide possible explanations for the occurrence of such results

  1. The correlation between GDP per capita and happiness score can be explained as an improvement in economic conditions may lead to higher levels of happiness;

  2. The correlation between healthy life expectancy and happiness score can be explained by the importance of physical health and longevity for personal well-being;

  3. The correlation between social support and happiness score can be explained as the positive impact of emotions and mutual assistance felt by people in social relationships and support systems on happiness.

````python
from scipy.stats import norm
from scipy import stats
````

````python
fig, ax = plt.subplots(2, 2, figsize=(8, 8))

sns.distplot(df['Score'], fit=norm, ax=ax[0, 0])
sns.distplot(df['GDP per capita'], fit=norm, ax=ax[0, 1])
sns.distplot(df['Healthy life expectancy'], fit=norm, ax=ax[1, 0])
sns.distplot(df['Social support'], fit=norm, ax=ax[1, 1])

print('Score:\nSkewness: {}\nKurtosis: {}\n'.format(
    df['Score'].skew(), df['Score'].kurt()
))
print('GDP per capita:\nSkewness: {}\nKurtosis: {}\n'.format(
    df['GDP per capita'].skew(), df['GDP per capita'].kurt()
))
print('Healthy life expectancy:\nSkewness: {}\nKurtosis: {}\n'.format(
    df['Healthy life expectancy'].skew(), df['Healthy life expectancy'].kurt()
))
print('Social Support:\nSkewness: {}\nKurtosis: {}\n'.format(
    df['Social support'].skew(), df['Social support'].kurt()
))
````
